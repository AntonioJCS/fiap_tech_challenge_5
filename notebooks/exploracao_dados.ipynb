## Modelagem e Normalizada dos dados para melhor visualização

#### Fonção que quebra o dados original em varias tabelas organizadas

```python
def estruturação_dos_dados(data: pd.DataFrame) -> dict[list]:
    dfs_dict = {}
    for con in data.columns:
        print(con)
        dfs_dict[con] = []

    for index, item in data.iterrows():
        for con in data.columns:
            if isinstance(item[con], dict):
                d = dict(item[con])
                d["codigo"] = index
                dfs_dict[con].append(d)
            elif isinstance(item[con], str):
                d = {con: item[con]}
                d["codigo"] = index
                dfs_dict[con].append(d)
            elif isinstance(item[con], list):
                d = item[con]
                for i in d:
                    i["codigo_vaga"] = index
                dfs_dict[con].extend(d)

    return dfs_dict
```

#### Função para contar e porcentagem de células vazias em uma coluna

```python
def calcular_nulos_na_coluna(df: pd.DataFrame, coluna: str):
    porcentagem_nulos = (df[coluna] == "").sum()
    total_de_linha = len(df)
    porcentagem_nulos = (porcentagem_nulos/total_de_linha) * 100
    print(f"Porcentagem de nulos entre {total_de_linha} linhas: {porcentagem_nulos:.2f}%")
```
